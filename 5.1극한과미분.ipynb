```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return 3*x**2 + 4*x -5

def my_func_dif(x):
    return 6*x + 4 # 도함수

x = np.linspace(-3,3)
y = my_func(x)

a = 1
y_t = my_func_dif(a)*x + my_func(a) - my_func_dif(a)*a
# x=1일때 접선, 접선 식을 사용

plt.plot(x,y,label="y")
plt.plot(x, y_t, label="y_t")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
